```python
import glob
filenames = []
for file in glob.glob("*.ict"):
    filenames.append(file)
print(filenames)
```

```python
file = filenames[2]
with open(file) as infile:
    text = []
    for line in infile:
        text.append(line)

lable = 'UTC,Latitude,Longitude,Temperature,Pointing_angle,AGL,Footprint_diameter\n'
c = 0
for i in text:
    if i == lable:
        print(c)
        count = c
    c = c+1
import pandas as pd
df = text[count:]
mat = [n.split(',') for n in df]

df = pd.DataFrame(mat)

new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df["UTC"] = pd.to_datetime(df["UTC"], unit='s')
df
```
